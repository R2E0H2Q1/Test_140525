```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
customers_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/customers.csv')
customers_df
```

```python
items_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/Items.csv')
items_df
```

```python
order_item_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/order_item.csv')
order_item_df
```

```python
orders_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/orders.csv')
orders_df
```

```python
# Data Analysis: Use the following data files: Customers.csv, Items.csv, order_item.csv, orders.csv

# 1. What is the average price of an item?
print('1. The Average for the prices is: ')
items_df[['item_price']].mean()
```

```python
# 2. Which customer has purchased the most products?

print('2. The Customer with more purchases is: ')
display(orders_df['customer_id'].value_counts().head(1))
```

```python
# Research and present in a graph the following details:
# Recommendation: Remove rows with empty values

print('1. Distribution of customers by gender: ')

plt.figure(figsize=(10, 4), dpi=200)
sns.countplot(data=customers_df, x='gender')
plt.xlabel('Customers per Gender')
```

```python
print('2. Distribution of purchase quantities by nationality: ')

purxnatio_df = pd.merge(order_item_df, customers_df, on='id', how='inner')
display(purxnatio_df)

sns.scatterplot(x='id', y='nationallity', data=purxnatio_df)
plt.xlabel('Customer ID')
plt.ylabel('Nationality')
plt.show()
```
